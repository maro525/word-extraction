```python
# -*- coding: utf-8 -*-
# 参考, http://tech.innovation.co.jp/2017/07/28/mecab.html

import re
import bs4
import sys
import MeCab
import urllib.request
```

```python
url = "http://www.sanspo.com/baseball/news/20180407/mlb18040711000002-n1.html"
soup = bs4.BeautifulSoup(urllib.request.urlopen(url).read(), "html.parser")
contents = soup.find(attrs={"name": re.compile(r'Description', re.I)}).attrs['content']
print(contents)
```

```python
output_words = []
m = MeCab.Tagger()
keywords = m.parse(contents)

for row in keywords.split("\n"):
    word = row.split("\t")[0]
    if word == "EOS":
        break
    else:
        pos = row.split("\t")[1].split(",")[0]
        if pos == "名詞":
            output_words.append(word)
for w in output_words:
    print(w)
```

```python
corpus1 = ["あい猫、う","あ、い、え"]
docs = [x for x in contents.split("。") if x !=""]
print(docs)
```

```python
count = CountVectorizer()

bag = count.fit_transform(docs)

# test
# X = count.fit_transform(corpus1)
# features = count.get_feature_names()
# print(features)
```

```python
# tfidf
tfidf = TfidfTransformer(use_idf=True, norm='l2', smooth_idf=True)
np.set_printoptions(precision=4)
cnt_voc = count.vocabulary_
tfidf_array = np.array(tfidf.fit_transform(bag).toarray())
```

```python
# ベクトルと単語を対応づけるための処理
vec = DictVectorizer()
vec.fit_transform(cnt_voc).toarray()
feature = vec.get_feature_names()
```

```python
threshold = 0.5
array = np.where(tfidf_array >= threshold)
print(array)
hit_word = np.unique(np.append(array[0], array[1]))
```

```python
for num in range(0, len(hit_word)):
    print(feature[hit_word[num]])
```
